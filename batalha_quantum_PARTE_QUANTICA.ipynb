# Problema de otimização

## Formulação quadrática

$$\begin{aligned}
\min_{x \in \{0, 1\}^n}  x^T \Sigma x - \mu^T x\\
\text{subject to: } A x = L \\
                    R^T x \leq c
\end{aligned}$$



- $x \in \{0, 1\}^n$ vetor binário que indica se o ativo foi escolhido ($x[i] = 1$) ou não ($x[i] = 0$),
- $\mu \in \mathbb{R}^n$ retorno médio dos ativos (`lista_produto.ret_med`),
- $\Sigma \in \mathbb{R}^{n \times n}$ matriz de coravariância (`matriz_de_covariancia`),
- $A$ Matriz do vínculo que define quantidade de produtos por classe,
- $L$ vetor com quantidade de produtos que devem ser selecionados de cada classe (`dict_quantidade_produtos`).
- $R$ matriz que representa a avaliação de risco dos produtos (`lista_produto.aval`).
- $c$ máximo risco permitido

Vamos carregar o problema e transforma-lo em seu analogo quantico com o modelo de ising, para que possamos resolver o problema usando computadores quanticos.

````python
# Carregando o problema
qubo_easy = carrega_problema_qubo(120)

# Transformando o problema em seu analogo do modelo de Ising, para ser resolvido no computador quantico
ising, offset = qubo_easy.to_ising()
````

Para podermos verificar as solucoes obtidas com os hardwares quanticos, vamos obter a solucao classica como uma referencia.

````python
# Resolvendo o problema classicamente, para podermos verificar as solucoes
numpy_solver = NumPyMinimumEigensolver()

result_qubo_easy = numpy_solver.compute_minimum_eigenvalue(qubo_easy.to_ising()[0])

answer_qubo_easy = np.array([int(char) for char in list(result_qubo_easy.eigenstate.to_dict().keys())[0][::-1]])

# A resposta com o metodo classico
answer_qubo_easy
````

Construindo o circuito quantico para realizar a construcao do Ansatz:

````python
import qiskit

# Definindo o numero de qubits e o numero de repeticoes
num_qubits = 6
reps = 1

# Criando o circuito quantico
circuit = qiskit.QuantumCircuit(num_qubits)

# Definindo o nome dos parametros dos angulos de rotacoes
parameter_names = [f"theta_{i}" for i in range(num_qubits)]

# Criando o vetor de parametros dos angulos
rotation_parameters = qiskit.circuit.ParameterVector("rotation_angles", length=6)

# Adicionado as layers de rotacao nos qubits
for _ in range(reps):
    for qubit in range(num_qubits):
        circuit.ry(rotation_parameters[qubit], qubit)

    # Adicionando CZ gates para obter o emaranhamento entre os qubits
    for i in range(num_qubits - 1):
        circuit.cz(i, i + 1)

# Visulizando o circuito
circuit.draw(output="mpl", style="clifford")
````

````python
circuit.depth()
````

O circuito possui 6 parametros e profundidade critica 6. Vamos agora gerar um ponto inicial para os angulos de rotacao do circuito e otimizar os parametros utilizando o VQE.

````python
initial_point = [1, 1, 1, 1, 1, 1]

vqe_solver = VQE(ansatz=circuit,
                 optimizer=COBYLA(maxiter=200),
                 initial_point=initial_point,
                 estimator = estimator)
````

Medindo o tempo de execucao para obter o resultado:

````python
import time

start_time = time.time()
resultado_vqe = vqe_solver.compute_minimum_eigenvalue(qubo_easy.to_ising()[0])
end_time = time.time()
execution_duration = end_time - start_time
print(f"Execution Time: {execution_duration} seconds")
````

Esse resultado esta de acordo com o Histograma obtido acima e com os resultados obtidos classicamente. Assim, sabemos que a resposta esta correta. Mas qual a taxa de erro? Podemos estimar o erro como

````python
def estimate_error_rate(counts, expected_outcome):
    total_shots = sum(counts.values())
    correct_shots = counts.get(expected_outcome, 0)
    error_rate = 1 - (correct_shots / total_shots)
    return error_rate

# Utilizando a solucao classica como resultado esperado
expected_outcome = '010101'
error_rate = estimate_error_rate(counts, expected_outcome)
print(f"Error Rate: {error_rate * 100:.2f}%")
````
